```python
# Program Pengelolaan Nilai Mahasiswa
import matplotlib.pyplot as plt

nilai = []

# Input 10 nilai
for i in range(10):
    n = int(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
    nilai.append(n)

# Menampilkan semua nilai
print("\nDaftar Nilai:", nilai)

# Nilai tertinggi & terendah
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)

print("Nilai Tertinggi:", nilai_tertinggi)
print("Nilai Terendah:", nilai_terendah)

# Rata-rata
rata_rata = sum(nilai) / len(nilai)
print("Rata-rata:", rata_rata)

# Jumlah mahasiswa lulus (>= 60)
lulus = 0
for n in nilai:
    if n >= 60:
        lulus += 1

print("Jumlah Mahasiswa Lulus:", lulus)
print("Jumlah Tidak Lulus:", len(nilai) - lulus)


nilai = [70, 55, 80, 90, 60, 40, 75, 85, 50, 65]

# Grafik nilai tertinggi & terendah
labels = ['Tertinggi', 'Terendah']
values = [max(nilai), min(nilai)]

plt.figure()
plt.bar(labels, values, color=['blue', 'red'])
plt.title("Perbandingan Nilai")
plt.show()8

# Grafik kelulusan
lulus = sum(1 for n in nilai if n >= 60)
tidak_lulus = len(nilai) - lulus

labels = ['Lulus', 'Tidak Lulus']
values = [lulus, tidak_lulus]

plt.figure()
plt.pie(values, labels=labels, autopct='%1.1f%%',
        colors=['blue', 'red'])
plt.title("Persentase Kelulusan")
plt.show()
```
